```python
def calcular_desconto(valor_original, percentual_desconto):
    valor_final = valor_original * (1 - percentual_desconto / 100)
    return valor_final
```

```python
calcular_desconto(200, 15)
```

```python
def encontrar_extremos(lista):
    maior = max(lista)
    menor = min(lista)
    return maior, menor
```

```python
encontrar_extremos([4, 7, 1, 9, 3])
```

Exercício 8

```python
def calculadora(a, b, operacao):
    if operacao == 'soma':
        return a + b
    elif operacao == 'subtracao':
        return a - b
    elif operacao == 'multiplicacao':
        return a * b
    elif operacao == 'divisao':
        if b != 0:
            return a / b
        else:
            return "Erro: Divisão por zero não é permitida."
    else:
        return "Operação inválida."
```

```python
calculadora(10, 2, 'divisao')
```

```python
calculadora(10, 2, 'multiplicacao')
```

```python
calculadora(10, 2, 'soma')
```

```python
calculadora(10, 2, 'subtracao')
```

exercício 10

```python
def avaliar_aluno(nota):
    if nota >= 7:
        return "Aprovado"
    elif 5 <= nota < 7:
        return "Recuperação"
    else:
        return "Reprovado"
```

```python
avaliar_aluno(6)
```

```python
avaliar_aluno(10)
```
