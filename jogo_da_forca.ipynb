```python
import random

def escolher_palavra():
    lista = ['python', 'programacao', 'escola', 'computador', 'internet']
    return random.choice(lista)

def mostrar_palavra(palavra, letras_cert):
    return " ".join([letra if letra in letras_cert else '_' for letra in palavra])

def tentar_letra(letra, palavra, letras_cert, letras_erradas):
    if letra in palavra:
        letras_cert.add(letra)
        return True
    else:
        letras_erradas.add(letra)
        return False

def verificar_vitoria(palavra, letras_cert):
    return set(palavra).issubset(letras_cert)

def jogo_da_forca():
    palavra = escolher_palavra()
    letras_cert = set()
    letras_erradas = set()
    tentativas = 6

    print("Bem-vindo ao jogo da forca!")
    while tentativas > 0:
        print(mostrar_palavra(palavra, letras_cert))
        print(f"Tentativas restantes: {tentativas}")
        print(f"Erros: {' '.join(letras_erradas)}")
        letra = input("Tente uma letra: ").lower()

        if letra in letras_cert or letra in letras_erradas:
            print("Você já tentou essa letra.")
            continue

        if tentar_letra(letra, palavra, letras_cert, letras_erradas):
            print("Acertou!")
        else:
            print("Errou!")
            tentativas -= 1

        if verificar_vitoria(palavra, letras_cert):
            print(f"Parabéns! Você venceu. A palavra era '{palavra}'.")
            return

    print(f"Você perdeu! A palavra era '{palavra}'.")
```

```python
jogo_da_forca()
```
